```python
#Importing all relevant libraries
import pandas as pd
import numpy as rp
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import datetime as dt
```

```python
#Load Dataset from the table using spark
df = spark.table('workspace.default.retail_sales_dataset')
rs = df.toPandas()
```

```python
#Create a graph (Sales (Total Amount) by Category - Matplotlib)
sales_by_category = rs.groupby('Product Category')['Total Amount'].sum()
plt.figure()
sales_by_category.plot(kind='bar')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
# Sales (Total Amount) Trend Over Time
rs['Date'] = pd.to_datetime(rs['Date'])
sales_trend = rs.groupby('Date')['Total Amount'].sum()
plt.figure()
sales_trend.plot()
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Seaborn - Sales (Total Amount) by Region (Age)
plt.figure()
sns.barplot(data=rs, x='Age', y='Total Amount')
plt.title('Sales by Age')
plt.show()
```
