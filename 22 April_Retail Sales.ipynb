```python
date_column = pd.DatetimeIndex(rs['Date'])
rs['Year'] = date_column.year
rs['Month'] = date_column.month
rs['Day'] = date_column.day
```

```python
display(rs)
```

```python
#Sta
rs.groupby('Month')['Total Amount'].sum()
```

```python
#Create a graph (Sales (Total Amount) by Category - Matplotlib)
sales_by_category = rs.groupby('Product Category')['Total Amount'].sum()
plt.figure()
sales_by_category.plot(kind='bar')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
# Sales (Total Amount) Trend Over Time
rs['Date'] = pd.to_datetime(rs['Date'])
sales_trend = rs.groupby('Date')['Total Amount'].sum()
plt.figure()
sales_trend.plot()
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Seaborn - Sales (Total Amount) by Region (Month)
plt.figure()
sns.barplot(data=rs, x='Month', y='Total Amount')
plt.title('Sales by Month')
plt.show()
```
